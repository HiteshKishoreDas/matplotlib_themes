```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data  = np.linspace(0,4*np.pi,num=1000)

y_data1 = np.sin(x_data)
y_data2 = np.cos(x_data)
y_data3 = np.sin(2*x_data)

x_data_scatter = np.linspace(0,4*np.pi,num=10)
y_data_scatter = np.random.uniform(size=np.size(x_data_scatter))
```

```python
def plot_test ():

    plt.plot(x_data,y_data1)
    plt.plot(x_data,y_data2)
    plt.plot(x_data,y_data3, linestyle='dashed')

    plt.scatter(x_data_scatter, y_data_scatter, marker='o')
```

```python
plt.style.use('plot_style.mplstyle')
plot_test()
```

```python
plt.style.use('plot_style_dark.mplstyle')
plt.style.use('dark_background')
plot_test()
```
